```python
import re
```

```python
grade_map = {
    'A+': 10, 'A': 9.5, 'A-': 9,
    'B+': 8.5, 'B': 8, 'B-': 7.5,
    'C+': 7, 'C': 6.5, 'C-': 6,
    'D+': 5.5, 'D': 5, 'D-': 4.5,
    'F': 2
}
```

```python
def convert_review_score(score):
    score = str(score).strip().upper()

    # Clean up trailing spaces
    score = re.sub(r'\s+', '', score)

    # 1. Letter grade
    if score in grade_map:
        return grade_map[score]
    if '/' in score:
        try:
            num, denom = score.split('/')
            num = float(num)
            denom = float(denom)
            if denom == 0: return np.nan
            return round((num / denom) * 10, 2)
        except:
            return np.nan
    try:
        val = float(score)
        if 0 <= val <= 10:
            return val
        elif 10 < val <= 100:
            return round((val / 100) * 10, 2)
        else:
            return np.nan
    except:
        return np.nan

        
```

```python
final_movies['review_score'] = final_movies['review_score'].apply(convert_review_score)
```

```python
pt = final_movies.pivot_table(index='movie_title', columns='critic_name', values='review_score')
```

```python
pt.fillna(0 , inplace = True)
```

```python
similarities = cosine_similarity(pt)
similarities.shape
```

```python
def recommend(movie):
    index = np.where(pt.index == movie)[0][0]
    movies_list = sorted(list(enumerate(similarities[index])),key = lambda x:x[1] , reverse = True)[1:6]
    for i in movies_list:
        print(pt.index[i[0]])
```

```python
gen_mov = final_movies[['rotten_tomatoes_link', 'movie_title' , 'movie_info' , 'genres' , 'directors' , 'actors']]
```

```python
gen_mov = gen_mov.drop_duplicates()
```

```python
gen_mov['actors'] = gen_mov['actors'].fillna('')
gen_mov['actors'] = gen_mov['actors'].apply(lambda x: " ".join(x.replace(" ", "").split()))
```

```python
gen_mov['directors'] = gen_mov['directors'].fillna('')
gen_mov['directors'] = gen_mov['directors'].apply(lambda x: " ".join(x.replace(" ", "").split()))
```

```python
def get_top_3_actors(actor_str):
    if isinstance(actor_str, str):
        return ",".join(actor_str.split(",")[:3])
    else:
        return ""

gen_mov['actors'] = gen_mov['actors'].apply(get_top_3_actors)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for word in text.split():
        y.append(ps.stem(word))
    return " ".join(y)
```

```python
gen_mov['tags'] = gen_mov['tags'].fillna('')
gen_mov['tags'] = gen_mov['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(gen_mov['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
gen_similarities = cosine_similarity(vectors)
gen_similarities
```

```python
sorted(list(enumerate(gen_similarities[0])) , reverse=True , key = lambda x:x[1])[1:6]
```

```python
def gen_recommend(name):
    movie_index = gen_mov[gen_mov['movie_title'] == name].index[0]
    distances = gen_similarities[movie_index]
    movie_list = sorted(list(enumerate(distances)) , reverse=True , key = lambda x:x[1])[1:6]

    for i in movie_list:
        print(gen_mov.iloc[i[0]].movie_title)
```

```python
def recommend_me(name):
    # Step 1: Content-based recommendation
    movie_index = gen_mov[gen_mov['movie_title'] == name].index[0]
    distances = gen_similarities[movie_index]
    content_based = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    recommended_titles = set()

    print("Recommendations for you:")
    for i in content_based:
        title = gen_mov.iloc[i[0]].movie_title
        if title not in recommended_titles:
            print(title)
            recommended_titles.add(title)

    print("\nYou may also like:")
    # Step 2: Collaborative filtering
    index = np.where(pt.index == name)[0][0]
    collab_based = sorted(list(enumerate(similarities[index])), key=lambda x: x[1], reverse=True)[1:]

    count = 0
    for i in collab_based:
        title = pt.index[i[0]]
        if title not in recommended_titles:
            print(title)
            recommended_titles.add(title)
            count += 1
            if count == 5:
                break
```

```python
recommend_me('John Wick')
```
